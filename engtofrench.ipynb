```python
from transformers import pipeline
import time
```

```python
translator = pipeline("translation_en_to_fr", model="Helsinki-NLP/opus-mt-en-fr")
```

```python
def dynamic_typing(text):
    """
    Prints the text one character at a time to simulate typing.
    """
    for char in text:
        print(char, end='', flush=True)
        time.sleep(0.05)  # 0.05 sec delay between characters
    print()  # Add new line at the end

# Step 5: Interactive chatbot loop
print("English-to-French Chatbot (type 'exit' to quit)")
while True:
    english_text = input("You: ")

    # Exit condition
    if english_text.lower() == "exit":
        print("Chatbot: Goodbye! 👋")
        break

    # Translate English input to French
    result = translator(english_text)
    french_text = result[0]['translation_text']

    # Display French translation with typing effect
    print("Chatbot (French): ", end="")
    dynamic_typing(french_text)
```
